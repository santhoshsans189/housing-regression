```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",300)
pd.set_option('display.max_rows', 500)
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train = train.drop(["Id","Alley","PoolQC","Fence","MiscFeature"], axis = 1)
```

```python
for col in ("BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","Electrical","FireplaceQu","GarageType","GarageFinish",'GarageQual','GarageCond'):
    train[col]=train[col].fillna(train[col].mode()[0])
```

```python
dummy = pd.get_dummies(train, drop_first = True )
```

```python
dummy.head()
```

```python
y = dummy["SalePrice"]
x = dummy.drop(["SalePrice"],axis = 1)
x = x.iloc[:,:].values
```

```python
y = y.values.reshape(-1, 1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(x_train,y_train)
```

```python
y_pred_lr = linear.predict(x_test)
```

```python
y_train
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score 
print("mean_absolute_error : ",mean_absolute_error(y_test, y_pred_lr))
print("mean_squared_error : ",mean_squared_error(y_test, y_pred_lr))
print("r2_score : ",r2_score(y_test, y_pred_lr))
```

```python
def smape(A, F):
    return 100/len(A) * np.sum(2 * np.abs(F - A) / (np.abs(A) + np.abs(F)))
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
smape(y_test ,y_pred_lr)
```

```python
rmse(y_test ,y_pred_lr)
```

```python
from sklearn.model_selection import cross_val_score
mse_lr = cross_val_score(linear,x,y,cv =10)
mean_mse = np.mean(mse_lr)
print(mean_mse)
```
